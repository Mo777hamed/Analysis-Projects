```python
df.sort_values('Date', inplace=True)
df['revenue_prev_period'] = df.groupby('Product line')['Revenue'].shift(1)
df['growth_rate'] = (df['Revenue'] - df['revenue_prev_period']) / df['revenue_prev_period'] * 100
```

```python
growth_rate = df.groupby('Product line')['growth_rate'].mean()
print(growth_rate)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
sales_over_time = df.groupby(['Date', 'Product line'])['Revenue'].sum().unstack()

sales_over_time.plot(kind='line', marker='o')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.title('Sales Trend over Time')
plt.legend(title='Product Line', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
